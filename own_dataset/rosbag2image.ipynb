```python
import numpy as np
import rosbag
from sensor_msgs.msg import Image
import cv2


def rosbag_to_img(bag, topic):
    image_times = []
    for topic, msg, t in bag.read_messages(topics= topic):
        try:
            dtype = np.dtype("uint8")  # Assuming image is uint8
            dtype = dtype.newbyteorder('>' if msg.is_bigendian else '<')
            image_data = np.fromstring(msg.data, dtype=dtype)

            # Assuming image is grayscale, if not, you have to handle the number of channels
            image_data = image_data.reshape(msg.height, msg.width, 3) 
            
            timestamp_secs = msg.header.stamp.secs + msg.header.stamp.nsecs * 1e-9
            image_times.append(timestamp_secs)

        except Exception as e:
            print(e)
        else:
            # Now image_data contains the image array that you can work with
            # For instance, you can display it with OpenCV
            cv2.imshow('image', image_data)
            cv2.waitKey(3)
    print(len(image_times),'Image timestamps:', image_times)
    cv2.destroyAllWindows()
    bag.close()


```

```python
bag = rosbag.Bag(r'/media/bru/DATA/BAG_FILES/19.01.23/data_mocap/linear/linear.bag')
topic = ['/camera/aligned_depth_to_color/image_raw']

rosbag_to_img(bag,topic)
```
